```python
df['budget'] = pd.to_numeric(df['budget'],errors = 'coerce')
df['gross'] = pd.to_numeric(df['gross'], errors ='coerce')
```

```python
df['budget']
df['gross']
```

```python
df['Release']= df['released'].astype(str).str[:]
df['Release']
```

```python
cat_cols = ['name','rating','genre','director','writer','country','company']  
df[cat_cols] = df[cat_cols].apply(lambda x: x.astype(str).str.strip())
```

```python
df[cat_cols]
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['budget'] = df['budget'].fillna(df['budget'].median())
df['budget']
```

```python
df['gross'] = df['gross'].fillna(df['gross'].median())
df['gross']
```

```python
df['rating'] = df['rating'].fillna('unknown')
df['rating']
```

```python
df['runtime'] = df['runtime'].fillna(df['runtime'].median())
df['runtime']
```

```python
df.isnull().sum()
```

# Budget vs Gross (scatter)

- understanding revenue relationships

```python
plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budjet vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('budget')
plt.show()
```

```python
import seaborn as sns
sns.regplot(x = 'budget', y ='gross', data = df, scatter_kws = {"color": "red"},line_kws = {"color" : "blue"})
plt.show()
```

```python
df[['year','score','votes','budget','gross','runtime']].corr()
```

# Correlation matrix for Numeric Features( heatmap)

```python
# high coorlection between budget and gross

correlation_matrix = df[['year','score','votes','budget','gross','runtime']].corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")
plt.show()
```

# score vs votes

- shows how audience votes relate to IMDB score

```python
plt.figure(figsize = (8,5))
sns.scatterplot(data= df, x = 'votes', y = 'score')
plt.show()
```

```python
top_gross = df['gross'].value_counts().head(10)
plt.figure(figsize = (8,5))
sns.barplot(x = top_gross.values, y = top_gross.index)
plt.show()
```

```python
plt.figure(figsize =(10,6))
sns.lineplot(data = df, x = 'year', y = 'gross')
plt.title('gross revenue by year')
plt.xlabel('year')
plt.ylabel('gross')
plt.show()
```

# country wise movie count

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.show()
```

```python
top_directors = df['director'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title("Top 10 Directors by Number of Movies")
plt.xlabel("Movies")
plt.ylabel("Director")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['rating'].value_counts().plot(kind='bar')
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Movie Rating Distribution")
plt.show()
```
